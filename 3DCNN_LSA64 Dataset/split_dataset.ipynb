```python
#Pad the Video ID with zeros to match the format in the dataset
lsa['VideoID'] = lsa['Video_ID'].apply('{:0>3}'.format)
lsa.head()
```

```python
import shutil
import os
```

```python

#Create the required directories for Video Dataset and to Split the videos into Train and test videos
def create_dir():
    if not os.path.exists("LSA_64"):
        os.mkdir("LSA_64")
    if not os.path.exists("LSA_64/train_videos"):
        os.mkdir("LSA_64/train_videos")
    if not os.path.exists("LSA_64/test_videos"):
        os.mkdir("LSA_64/test_videos")

create_dir()
```

```python
#Create the required directories for Video Dataset and to Split the videos into Train and test videos
def create_train():
    for i in range(len(lsa)):
        if not os.path.exists("LSA_64/train_videos/" + lsa['Word'][i]):
            os.mkdir("LSA_64/train_videos/" + lsa['Word'][i])
    
create_train()
```

```python
#Create the required directories for Video Dataset and to Split the videos into Train and test videos
def create_test():
    for i in range(len(lsa)):
        if not os.path.exists("LSA_64/test_videos/" + lsa['Word'][i]):
            os.mkdir("LSA_64/test_videos/" + lsa['Word'][i])
    
create_test()
```

```python
source = "C:/Users/LEGION/Jupyter/LSA64_3DCNN/dataset/LSA64_Cropped/"
train_dest = "LSA_64/train_videos/"
test_dest = "LSA_64/test_videos/"


per_id = ['001', '002','003', '004', '005', '006', '007', '008', '009', '010']
Vid_id = ['001', '002','003', '004', '005']
        
def copy_train_videos():
    for i in range(len(lsa)):
        dest = train_dest + lsa['Word'][i] + "/"
        for file in myList:
            if((file[0] == lsa['VideoID'][i]) & (file[2].split('.')[0] in Vid_id)):
                filename = '_'.join(file)
                src = source + str(filename)
                shutil.copy(src, dest)
            

copy_train_videos()
```

```python
def copy_test_videos():
    for i in range(len(lsa)):
        dest = test_dest + lsa['Word'][i] + "/"
        for file in myList:
            if((file[0] == lsa['VideoID'][i]) & (file[2].split('.')[0] not in Vid_id)):
                filename = '_'.join(file)
                src = source + str(filename)
                shutil.copy(src, dest)
            

copy_test_videos()
```
